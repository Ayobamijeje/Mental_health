### Stable Diffusion - Image generation 

```python
# Import  modules 
from tensorflow import keras
import keras_cv
import matplotlib.pyplot as plt

```

```python
#Initiating the SD model

img_generator = keras_cv.models.StableDiffusion(
    img_width=512, 
    img_height=512, 
    jit_compile=False
)

```

```python
prompt = 'a cat  and a mouse'

images = img_generator.text_to_image(prompt, batch_size=1)
```

```python
plt.imshow(images[0])
plt.axis()
```

```python
prompt = 'a cat chasing a mouse'

images = img_generator.text_to_image(prompt, batch_size=1)
```

```python
plt.imshow(images[0])
plt.axis()
```
